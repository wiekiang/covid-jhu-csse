## Manipulating Data

```
# re-assign method
covid_daily_report["Lat"] = covid_daily_report["Lat"].fillna(0)
covid_daily_report["Long_"] = covid_daily_report["Long_"].fillna(0)
covid_daily_report["Confirmed"] = covid_daily_report["Confirmed"].fillna(0)
covid_daily_report["Recovered"] = covid_daily_report["Recovered"].fillna(0)

# inplace method
covid_daily_report["Incident_Rate"].fillna(0, inplace=True)
covid_daily_report["People_Tested"].fillna(0, inplace=True)
covid_daily_report["People_Hospitalized"].fillna(0, inplace=True)
covid_daily_report["Mortality_Rate"].fillna(0, inplace=True)
covid_daily_report["Testing_Rate"].fillna(0, inplace=True)
covid_daily_report["Hospitalization_Rate"].fillna(0, inplace=True)
```

```
covid_daily_report_drop_missing = covid_daily_report.dropna()
covid_daily_report_drop_missing.info()
```

```
covid_daily_report_drop_missing
```

```
covid_daily_report_drop_missing.to_csv("covid_backup_drop_missing.csv")
```

```
covid_daily_report["Recovered_Percentage"] = covid_daily_report["Recovered"] / covid_daily_report["Confirmed"] * 100
covid_daily_report
```

```
covid_daily_report.drop("Recovered_Percentage", axis=1, inplace=True)
```

```
covid_daily_report_random = covid_daily_report.sample(frac=1)
covid_daily_report_random
```

```
# Select only 20% of data
covid_daily_report_random.sample(frac=0.2)
```

```
covid_daily_report_random.reset_index(drop=True, inplace=True)
covid_daily_report_random
```

```
covid_daily_report["Active"] = 
    covid_daily_report["Active"].apply(lambda x: x + covid_daily_report["Deaths"])
covid_daily_report
```

```
covid_daily_report["Active"] = covid_daily_report["Active"].apply(lambda x: x - covid_daily_report["Deaths"])
covid_daily_report
```
